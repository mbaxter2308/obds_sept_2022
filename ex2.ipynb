# Assign column names

```python
gtf.columns=['seqname', 'source' , 'feature' , 'start' , 'end', 'score' , 'strand' ,  'frame' ,  'attribute' ]
                          
```

```python
gtf.head()
```

# Subset all values from ‘ensembl_havana’ and assign to new df called ‘ens_df’ 

```python
ens_gtf = gtf[gtf["source"].str.contains("ensembl_havana")]
ens_gtf
```

# What is the size of ens_df? 

```python
ens_gtf.shape
```

# Group ens_df by ‘feature’ column & apply the ‘count’ function to get the number of rows in each feature

```python
(ens_gtf.groupby("feature")
   .count() 
)
```

# Create a new df called ‘genes’ by subsetting all rows with ‘gene’ as a feature 

```python
genes = ens_gtf[ens_gtf["feature"].str.contains("gene")]
```

```python
genes
```

# Create a new column called ‘gene_length’ and calculate the length of each gene

```python
genes["gene_length"] = genes["end"] - genes["start"]
genes
```

```python
genes = genes.assign(gene_length = genes['end'] - genes['start'], 
                    )
genes
```

# Calculate the mean and median of gene lengths 

```python
genes.describe()
```

# Find the top 10 longest genes 

```python
genes.sort_values(by=['gene_length'], ascending=False).head(10)
```

# Find the lowest 20 shorest genes 

```python
genes.sort_values(by=['gene_length']).head(20)
```
